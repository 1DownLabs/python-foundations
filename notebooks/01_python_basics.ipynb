```python
# Using a loop, calculate the sum of numbers from 1 to 100.


#Using For loop:
n = 0 

for a in range (1,101):
    n += a 

print (n)


# Using While Loop

Number = 0 
count = 1 
limit = 100 
while count <= limit:
    Number += count
    count += 1

print (f"Total Sum is {Number}")
```
