```python
# Variables and Types

x = 10 
y = 3.5
name = "Martin" 


print(type(x))
print(type(y))
print(type(name))
```

```python
# Basic Maths

a = 2 
b = 3 


print("Sum: ",a + b)
print("Division: ",b/a)
print("Multiplication: ", a*b)
print("Power: ",b**a)
```

```python
# Strings

brand_name = "1DownLabs"

print(brand_name.upper())
print(brand_name.lower())
print(f"Name of the Brand is {brand_name}")
```

```python
# Lists

Names = ["Mumbai", "Delhi", "New York", "Jersey City", "Boston"]

print("Name at 0th Location is: ",Names[0])


New_name = input("Name you want to add?")
Names.append(New_name)
print(Names)


```
