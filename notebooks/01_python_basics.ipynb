## Lists

```python
Names = ["Dalls", "San Antonio", "New York", "Jersey City", "Boston"]

print("Name at 0th Location is: ",Names[0])

#Append New city in the list
Names.append("Miami")
print(Names)


```

```python
#last 3 elements of list
print(Names[len(Names)-3: len(Names)])
print(Names[-3:])
```

```python
#First 3 elements of list
print(Names[:3])
```

```python
#Slice list from 2 onwards
print(Names[2:])
```

```python
Names[2] = "Philidephia"
```

```python
Names[2] = Names.append("New York")
```
