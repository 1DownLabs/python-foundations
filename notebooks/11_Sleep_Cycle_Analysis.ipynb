```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/arpitshukla/1DownLabs/data-science/python-foundations/data/raw/Smartphone_Usage_Productivity_Dataset_50000.csv")

df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
total_by_age = df.groupby('age', as_index=False)[['daily_phone_hours', 'social_media_hours', 'work_productivity_score', 'sleep_hours', 'sleep_hours']].mean()
```

```python
# Averag daily phone hours by Age 

plt.plot(total_by_age['age'], total_by_age['daily_phone_hours'], alpha=1)
plt.xlabel('Age')
plt.ylabel('Avg daily phone hours')
plt.show()
```

```python
# Averag social Media hours by Age 

plt.plot(total_by_age['age'], total_by_age['social_media_hours'], alpha=1)
plt.xlabel('Age')
plt.ylabel('Avg social Media hours')
plt.show()
```

```python
# avg work productivity score by age
plt.plot(total_by_age['age'], total_by_age['work_productivity_score'], alpha=1)
plt.xlabel('Age')
plt.ylabel('Avg work_productivity_score')
plt.show()
```
