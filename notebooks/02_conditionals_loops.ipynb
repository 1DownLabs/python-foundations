```python
# Comparison and Boolean
a = 10
b = 20

print(a == b)
print(a != b)
print(a < b)
print(a >= b)
```

```python
# Boolean logic (AND / OR / NOT)
is_senior = True
knows_python = False

print(is_senior and knows_python)
print(is_senior or knows_python)
print(not knows_python)
```

```python
# if / elif / else (simple)
score = 75

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 60:
    print("C")
else:
    print("Fail")
```

```python
# for loop basics
projects = ["python-foundations", "data-cleaning-lab", "pandas-case-studies"]

for p in projects:
    print("Project:", p)
```

```python
# for loop with range
for i in range(1, 6):
    print(i)
```

```python
# while loop basics
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

```python
# break / continue
nums = [1, 2, 3, 4, 5, 6, 7]

for n in nums:
    if n == 5:
        break
    print("n:", n)
```

```python
for n in nums:
    if n % 2 == 0:
        continue
    print("odd:", n)
```
